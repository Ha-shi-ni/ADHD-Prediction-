```python

# Columns of interest
cols = [
    'Laterality Index (Score)',
    'Color Vision Test Score',
    'Corporal Punishment Score',
    'Inconsistent Discipline Score',
    'Involvement Score',
    'Other Discipline Practices Score',
    'Poor Monitoring Score',
    'Positive Parenting Score',
    'Conduct Problems Scale',
    'Total Difficulties Score',
    'Emotional Problems Scale',
    'Externalizing Score',
    'Generating Impact Score',
    'Hyperactivity Scale',
    'Internalizing Score',
    'Peer Problems Scale',
    'Prosocial Scale',
    'Age at Time of MRI Scan'
]

# Include the ADHD column and the other columns
subset_df = manipulated_df[['ADHD_Outcome'] + cols]

# Calculate correlation of ADHD with each column
correlations = subset_df.corr()['ADHD_Outcome'].drop('ADHD_Outcome')

print("Correlation of ADHD with each specified variable:\n")
print(correlations)
```

```python
import pandas as pd
from scipy.stats import pearsonr
from statsmodels.stats.multitest import multipletests  # optional, for p‑value correction

# --- 1. Compute r and p for every column --------------------------
results = []
for col in cols:                                   # <- your list from above
    r, p = pearsonr(subset_df['ADHD_Outcome'], subset_df[col])

    # --- 2. Classify strength (Cohen-style thresholds) ------------
    abs_r = abs(r)
    if   abs_r >= 0.90: strength = "very strong"
    elif abs_r >= 0.70: strength = "strong"
    elif abs_r >= 0.50: strength = "moderate"
    elif abs_r >= 0.30: strength = "weak"
    else:               strength = "very weak"

    results.append({'Variable': col, 'r': r, 'p_value': p, 'strength': strength})

results_df = pd.DataFrame(results)

# --- 3. *Optional*: adjust p‑values for multiple tests ------------
# Here we apply Benjamini–Hochberg (FDR), but you can switch to 'bonferroni'
adj = multipletests(results_df['p_value'], method='fdr_bh')
results_df['p_adj'] = adj[1]                       # adjusted p
results_df['significant?'] = results_df['p_adj'] < 0.05

# --- 4. Pretty‑print ---------------------------------------------
results_df = (results_df
              .set_index('Variable')
              .sort_values('p_value'))             # or sort by |r|
print(results_df.round(4))
```

```python
class GCNNet(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=1):
        super(GCNNet, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = self.fc(x)
        return torch.sigmoid(x)
```

```python
from tensorflow.keras import layers, models, callbacks

def build_mlp(input_dim, units=[256,128,64], dropout=0.3):
    inp = layers.Input(shape=(input_dim,))
    x = layers.BatchNormalization()(inp)
    for u in units:
        x = layers.Dense(u, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(dropout)(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    model = models.Model(inp, out)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

mlp_model = build_mlp(X_train_s.shape[1])

# Early stopping
es = callbacks.EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Train
history = mlp_model.fit(
    X_train_s, y_train,
    validation_data=(X_val_s, y_val),
    epochs=200, batch_size=32,
    class_weight=cw_dict,
    callbacks=[es],
    verbose=2
)
```

** Hybrid Deep Learning Model (Behavioral + Brain)**

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input, Model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# --- 1️⃣ Separate feature groups ---
behavioral_cols = [
    'SDQ_SDQ_Hyperactivity',
    'SDQ_SDQ_Externalizing',
    'SDQ_SDQ_Difficulties_Total',
    'SDQ_SDQ_Generating_Impact',
    'SDQ_SDQ_Internalizing',
    'SDQ_SDQ_Conduct_Problems',
    'SDQ_SDQ_Peer_Problems'
]

brain_cols = [col for col in merged_df.columns if 'throw_' in col]  # all PCA features

y = merged_df['ADHD_Outcome'].values
X_behavioral = merged_df[behavioral_cols].values
X_brain = merged_df[brain_cols].values


# --- 2️⃣ Split data ---
Xb_train, Xb_test, Xp_train, Xp_test, y_train, y_test = train_test_split(
    X_behavioral, X_brain, y, test_size=0.2, random_state=42
)

# --- 3️⃣ Define two input branches ---

# Behavioral branch
input_behavior = Input(shape=(Xb_train.shape[1],), name="behavioral_input")
x1 = layers.Dense(64, activation='relu')(input_behavior)
x1 = layers.Dropout(0.3)(x1)
x1 = layers.Dense(32, activation='relu')(x1)

# Brain (PCA) branch
input_brain = Input(shape=(Xp_train.shape[1],), name="brain_input")
x2 = layers.Dense(128, activation='relu')(input_brain)
x2 = layers.Dropout(0.3)(x2)
x2 = layers.Dense(64, activation='relu')(x2)

# Merge both branches
combined = layers.concatenate([x1, x2])
z = layers.Dense(32, activation='relu')(combined)
z = layers.Dropout(0.3)(z)
output = layers.Dense(1, activation='sigmoid')(z)

# --- 4️⃣ Build and compile model ---
model = Model(inputs=[input_behavior, input_brain], outputs=output)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# --- 5️⃣ Train model ---
history = model.fit(
    [Xb_train, Xp_train], y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

# --- 6️⃣ Evaluate on test set ---
test_loss, test_acc = model.evaluate([Xb_test, Xp_test], y_test)
print(f"\n✅ Test Accuracy: {test_acc:.4f}")

# --- 7️⃣ Plot training results ---
plt.figure(figsize=(10,5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training vs Validation Accuracy')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training vs Validation Loss')
plt.show()
```
